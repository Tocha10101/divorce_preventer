```python
divorce_data1

all_series = [list(divorce_data1[column]) for column in divorce_data1.columns]
max_values = [max(seria) for seria in all_series] # w skali od 0 do 4

for i in range(len(all_series)):
    for j in range(len(all_series[i])):
        all_series[i][j] /= max_values[i]
        


divorce_data = pd.DataFrame(data=all_series)
divorce_data = divorce_data.transpose()
divorce_data.columns = divorce_data1.columns
divorce_data
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
def make_test():
    features_X, labels_y = divorce_data.drop(['Class'], axis=1), divorce_data['Class']

    features_train, features_test, labels_train, labels_test = train_test_split(features_X, labels_y, test_size=0.33)
    #print(len(features_test))
    rf_classifier = RandomForestClassifier(random_state=1)

    rf_classifier.fit(features_train, labels_train)

    pred = rf_classifier.predict(features_test)

    accuracy = accuracy_score(labels_test, pred)
    return accuracy

accs = [make_test() for i in range(100)]
print(sum(accs) / len(accs))
```
